```python
''' 爬虫1============获取猫眼评分最好的100部电影'''

import pandas as pd
import numpy as np
from  bs4  import BeautifulSoup
import requests, json
from requests.exceptions import RequestException
import re

headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.75 Safari/537.36'}
def get_url():
    url = 'https://maoyan.com/board/4?offset=0'
    res = requests.get(url,headers=headers)
    if res.status_code == 200:
        resp = res.text
    soup = BeautifulSoup(resp, 'lxml')
    pages = [j.text for j in soup.find_all('a',class_=re.compile('page_'))][1:-1]
    #print(pages)
    page = [(int(i)-1)*10 for i in pages]
    url_a = 'https://maoyan.com/board/4?offset='
    urls = [url_a+str(i) for i in range(10)]
    return urls

def get_data():
    urls = get_url()
    df = []
    for url in urls :
        res = requests.get(url,headers=headers)
        if res.status_code == 200:
            resp = res.text
        soup = BeautifulSoup(resp, 'lxml')
        names = soup.find_all('p',class_='name')
        stars = soup.find_all('p', class_='star')
        showTimes = soup.find_all('p', class_=re.compile('releaset'))
        scores = soup.find_all('p', class_=re.compile('score'))
        for i in range(len(names)):
            name = names[i].string
            star = [j for j in stars[i].stripped_strings]
            showTime = [j for j in showTimes[i].stripped_strings]
            score = scores[i].text
            df.append([name, star, showTime, score])
    return df
def deal_df(df):
    dfs = pd.DataFrame(df)
    dfs.columns=['name','star','time','score']
    dfs['star'] = [s[0].replace('主演：','')for s in dfs['star']]   #   做了一点点数据的处理
    dfs['time'] = [s[0].replace('上映时间：','')for s in dfs['time']]
    return dfs

df = get_data()

df
```
